```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
iris = datasets.load_iris()
X = iris.data[:, :2] 
y = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
svm_rbf = SVC(kernel='rbf', C=80, gamma=100)
svm_poly = SVC(kernel='poly', degree=6 , C=50)
svm_rbf1 = SVC(kernel='rbf', C=50, gamma=35)
svm_poly1 = SVC(kernel='poly', degree=2 , C=2)
svm_rbf2 = SVC(kernel='rbf', C=1, gamma=0.1)
svm_poly2 = SVC(kernel='poly', degree=3 , C=1)
```

```python
svm_rbf.fit(X_train, y_train)
svm_poly.fit(X_train, y_train)
svm_rbf1.fit(X_train, y_train)
svm_poly1.fit(X_train, y_train)
svm_rbf2.fit(X_train, y_train)
svm_poly2.fit(X_train, y_train)
```

```python
y_pred_rbf = svm_rbf.predict(X_test)
y_pred_poly = svm_poly.predict(X_test)
y_pred_rbf1 = svm_rbf1.predict(X_test)
y_pred_poly1 = svm_poly1.predict(X_test)
y_pred_rbf2 = svm_rbf2.predict(X_test)
y_pred_poly2 = svm_poly2.predict(X_test)

acc_rbf = accuracy_score(y_test, y_pred_rbf)
acc_poly = accuracy_score(y_test, y_pred_poly)
acc_rbf1 = accuracy_score(y_test, y_pred_rbf1)
acc_poly1 = accuracy_score(y_test, y_pred_poly1)
acc_rbf2 = accuracy_score(y_test, y_pred_rbf2)
acc_poly2 = accuracy_score(y_test, y_pred_poly2)

print('Precisión del modelo RBF 1:', acc_rbf)
print('Precisión del modelo polinomial 1:', acc_poly)
print('Precisión del modelo RBF 2:', acc_rbf1)
print('Precisión del modelo polinomial 2:', acc_poly1)
print('Precisión del modelo RBF 3:', acc_rbf2)
print('Precisión del modelo polinomial 3:', acc_poly2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xx, yy = np.meshgrid(np.arange(X[:, 0].min() - 0.5, X[:, 0].max() + 0.5, 0.02),
                     np.arange(X[:, 1].min() - 0.5, X[:, 1].max() + 0.5, 0.02))
Z_rbf= svm_rbf.predict(np.c_[xx.ravel(), yy.ravel()])
Z_rbf= Z_rbf.reshape(xx.shape)
Z_poly = svm_poly.predict(np.c_[xx.ravel(), yy.ravel()])
Z_poly= Z_poly.reshape(xx.shape)

Z_rbf1= svm_rbf1.predict(np.c_[xx.ravel(), yy.ravel()])
Z_rbf1= Z_rbf1.reshape(xx.shape)
Z_poly1= svm_poly1.predict(np.c_[xx.ravel(), yy.ravel()])
Z_poly1= Z_poly1.reshape(xx.shape)

Z_rbf2= svm_rbf2.predict(np.c_[xx.ravel(), yy.ravel()])
Z_rbf2= Z_rbf2.reshape(xx.shape)
Z_poly2 = svm_poly2.predict(np.c_[xx.ravel(), yy.ravel()])
Z_poly2= Z_poly2.reshape(xx.shape)
```

```python

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].contourf(xx, yy, Z_rbf, alpha=0.4)
ax[0].scatter(X[:, 0], X[:, 1], c=y, s=20, edgecolor='k')
ax[0].set_title('Modelo RBF')


ax[1].contourf(xx, yy, Z_poly, alpha=0.4)
ax[1].scatter(X[:, 0], X[:, 1], c=y, s=20, edgecolor='k')
ax[1].set_title('Modelo polinomial')

for a in ax:
    a.set_xlabel('Longitud del sépalo [cm]')
    a.set_ylabel('Ancho del pétalo [cm]')
    
plt.show()
```
