```python
U, S, VT = svd(gris)
```

```python
print(np.shape(U))
print(np.shape(S))
print(np.shape(VT))
```

```python
Sigma = np.zeros((gris.shape[0], gris.shape[1]))
```

```python
gris.shape[0]
```

```python
Sigma[:gris.shape[0], :gris.shape[0]]= diag(S)
```

```python
np.shape(Sigma)
```

```python
n_elements = 50

# reconstrucci√≥n
Sigma = Sigma [:,:n_elements]
VT = VT[:n_elements,:]
B = U.dot(Sigma.dot(VT))

print(np.shape(imagen))
gris = rgb2gray(imagen)
plt.figure(figsize=(10,10))

plt.imshow(B, cmap = 'gray')
```

```python
gris = rgb2gray(imagen)

U, S, V = np.linalg.svd(gris)

n_elements = [159, 100, 50, 10, 1]

fig, axs = plt.subplots(nrows=1, ncols=len(n_elements), figsize=(100, 100))

for i, n in enumerate(n_elements):
 
    reconstruccion = np.dot(U[:, :n], np.dot(np.diag(S[:n]), V[:n, :]))
    axs[i].imshow(reconstruccion, cmap='gray')
    axs[i].set_title(f'n={n}', fontsize = 70)


plt.show()

```

"Conclusio"
Al disminuir el n√∫mero de elementos (n) utilizados en la reconstrucci√≥n de la imagen, estamos reduciendo la cantidad de informaci√≥n utilizada para crear la aproximaci√≥n. Por lo tanto, la calidad de la imagen reconstruida disminuir√° a medida que se reduzca la cantidad de informaci√≥n utilizada.

Aunque podemos reducir el tama√±o de los datos mediante la reducci√≥n del rango, esta disminuci√≥n de tama√±o viene con una p√©rdida de informaci√≥n, lo que puede traducirse en una disminuci√≥n de la calidad de la imagen.  La p√©rdida de informaci√≥n que se puede ser mala o no, dependiendo del objetivo que se tenga en mente. En algunos casos, puede ser beneficiosa al eliminar detalles insignificantes y simplificar el an√°lisis de la imagen, mientras que en otros casos, la p√©rdida de informaci√≥n puede ser cr√≠tica.
