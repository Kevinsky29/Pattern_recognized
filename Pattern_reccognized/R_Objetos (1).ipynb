#Ejemplo en clase

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
template = cv.imread('/content/Image20230309100626.png')
template = cv.cvtColor(template, cv.COLOR_BGR2GRAY)
```

```python
img = cv.imread('/content/Image20230309100558.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
img2 = img.copy()

cv2_imshow(img)
```

```python
cv2_imshow(template)
```

```python
w, h = template.shape[::-1]
res = cv.matchTemplate(img, template, cv.TM_CCOEFF_NORMED)
threshold = 0.8
loc = np.where(res >= threshold)

for pt in zip(*loc[::-1]):
    cv.rectangle(img, pt, (pt[0] + w, pt[1] + h), 255, 2)
    
plt.figure(figsize = (15,15))
plt.imshow(img, cmap = 'gray')
plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
template = cv.imread('/content/image_pacman_ghost.png')
template = cv.cvtColor(template, cv.COLOR_BGR2GRAY)
```

```python
cv2_imshow(template)
```

```python
img = cv.imread('/content/image_pacman.png')
img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
img2 = img.copy()

cv2_imshow(img)
```

```python
img_points = img.copy()
```

```python
w, h = template.shape[::-1]
res = cv.matchTemplate(img, template, cv.TM_CCOEFF_NORMED)
threshold = 0.9
loc = np.where(res >= threshold)

for pt in zip(*loc[::-1]):
    cv.rectangle(img, pt, (pt[0] + w, pt[1] + h), 255, 2)

plt.figure(figsize = (15,15))
plt.imshow(img, cmap = 'gray')
plt.show()
```

Si bajamos el umbral, podremos notar como los fantasmas seran detectados mas facil, esto debido a su similitud, asumo que con el umbral alto, la diferencia esta en el color de los fantasmas.
