#Opcion 1 rellenar datos con media aritmentica

```python
# Reemplaza los valores NaN con la media aritmética de la columna
df1 = df.fillna(df.mean())
df1.head()
```

```python
df1 = df1.fillna(df1.mean())

df1.head()
```

# Opcion 2 Remplazar los Nan con los valores de Mediana

```python
# Reemplaza los valores NaN con la mediana de la columna
df2 = df1.fillna(df1.median())
df2.head()
```

```python
nulos_por_columna = df2.isna().sum()
print(nulos_por_columna)
```

#Opcion 3 Remplazar los Nan con los valores mas frecuentes

```python
nulos_por_columna = df3.isna().sum()
print(nulos_por_columna)
```
